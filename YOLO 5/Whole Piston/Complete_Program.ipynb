```
import os
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.metrics import confusion_matrix, classification_report, roc_curve, auc

# ========= CONFIG =========
results_dir = "/content/drive/MyDrive/YOLOv5_Project/EVAL_RESULTS"
os.makedirs(results_dir, exist_ok=True)

# ========= PREPARE DATA =========
all_probs = np.array(all_probs)
y_true_probs = np.array(y_true_probs)
y_pred = np.argmax(all_probs, axis=1)

# ========= 1. Confusion Matrix =========
cm = confusion_matrix(y_true_probs, y_pred)
print("Confusion Matrix:\n", cm)

plt.figure(figsize=(6, 5))
sns.heatmap(cm, annot=True, fmt='d', cmap="Blues",
            xticklabels=["defect_set", "good_set"],
            yticklabels=["defect_set", "good_set"])
plt.xlabel("Predicted")
plt.ylabel("Actual")
plt.title("Confusion Matrix Heatmap")
plt.savefig(f"{results_dir}/confusion_matrix.png", dpi=300, bbox_inches='tight')
plt.show()

# ========= 2. Classification Report =========
report = classification_report(y_true_probs, y_pred, target_names=["defect_set", "good_set"])
print("\nClassification Report:\n", report)

with open(f"{results_dir}/classification_report.txt", "w") as f:
    f.write("=== Classification Report ===\n\n")
    f.write(report)

# ========= 3. Sigmoid Curve =========
x = np.linspace(-6, 6, 200)
sigmoid = 1 / (1 + np.exp(-x))

plt.figure(figsize=(7, 4))
plt.plot(x, sigmoid, color='blue', label="Sigmoid Function")
plt.axhline(0.5, color='red', linestyle='--', label="Decision Boundary (p=0.5)")
plt.title("Logistic Regression Sigmoid Curve")
plt.xlabel("Logit (z)")
plt.ylabel("Probability")
plt.legend()
plt.savefig(f"{results_dir}/sigmoid_curve.png", dpi=300, bbox_inches='tight')
plt.show()

# ========= 4. Probability Distribution =========
plt.figure(figsize=(7, 4))
plt.hist(all_probs[:, 0], bins=20, alpha=0.6, label="Defect Probabilities", color='red')
plt.hist(all_probs[:, 1], bins=20, alpha=0.6, label="Good Probabilities", color='green')
plt.axvline(0.5, color='black', linestyle='--', label="Threshold (0.5)")
plt.title("Model Prediction Probability Distribution")
plt.xlabel("Predicted Probability")
plt.ylabel("Frequency")
plt.legend()
plt.savefig(f"{results_dir}/probability_distribution.png", dpi=300, bbox_inches='tight')
plt.show()

# ========= 5. ROC Curve =========
y_score = all_probs[:, 1]  # Probability of good_set
fpr, tpr, _ = roc_curve(y_true_probs, y_score)
roc_auc = auc(fpr, tpr)

plt.figure(figsize=(6, 5))
plt.plot(fpr, tpr, color='blue', lw=2, label=f"ROC curve (AUC = {roc_auc:.2f})")
plt.plot([0, 1], [0, 1], color='gray', linestyle='--')
plt.xlabel("False Positive Rate")
plt.ylabel("True Positive Rate")
plt.title("ROC Curve")
plt.legend()
plt.savefig(f"{results_dir}/roc_curve.png", dpi=300, bbox_inches='tight')
plt.show()

print(f"\n✅ All results saved in: {results_dir}")
```
