```python
from ultralytics import YOLO
dataset_path = "/content/drive/MyDrive/Datasets"
save_dir = "/content/drive/MyDrive/yolo11_runs"

# Load YOLOv11 classification model (this should auto-download the weights)
model = YOLO("yolo11n-cls.pt")

# Print model info
model.info()
```

```python
results = model.train(
    data=dataset_path,     # Path to dataset (train/ and val/)
    epochs=50,             # You can increase epochs for better accuracy
    imgsz=224,             # Image size (224 is good for classification)
    batch=16,              # Adjust batch size depending on GPU memory
    project=save_dir,      # This ensures results are saved in Drive
    name="good_vs_defect"  # Subfolder name inside save_dir
)
```

```python
metrics = model.val(project=save_dir, name="good_vs_defect_val")
```
